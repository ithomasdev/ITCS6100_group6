## Data Analysis/Machine Learning

#### KMeans Clustering

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.impute import SimpleImputer
import pandas as pd

# read dataframe from csv
df = pd.read_csv("./data/athletes.csv")

# Drop rows with missing data
df = df.dropna()

# Encode the gender column
gender_encoded = pd.get_dummies(df['gender'], prefix='gender')

# Encode the name column
name_encoded = pd.get_dummies(pd.Categorical(df['name'], categories=df['name'].unique()), prefix='name')

# Select the columns for clustering
cluster_cols = ['gender', 'age', 'height', 'weight', 'fran', 'helen', 'grace', 'filthy50', 'fgonebad',
                'run400', 'run5k', 'candj', 'snatch', 'deadlift', 'backsq', 'pullups']

# Scale the numeric columns
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[cluster_cols[1:]])

# Concatenate the encoded gender, name, and scaled numeric columns
cluster_data = pd.concat([gender_encoded, name_encoded, pd.DataFrame(scaled_data, columns=cluster_cols[1:])], axis=1)

# create an imputer with mean strategy
imputer = SimpleImputer(strategy='mean')

# Fit the imputer object to the data
imputer.fit(cluster_data)

# Impute the missing values
cluster_data_imputed = imputer.transform(cluster_data)

# Perform clustering using K-means
kmeans = KMeans(n_clusters=3, n_init=10, random_state=0)
kmeans.fit(cluster_data_imputed)

# Evaluate clustering performance using silhouette score
labels = kmeans.labels_
silhouette_avg = silhouette_score(cluster_data_imputed, labels)
print("Silhouette score: ", silhouette_avg)
```

```python
import matplotlib.pyplot as plt

# Obtain the cluster centers
centers = kmeans.cluster_centers_

# Plot the cluster centers
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.5)
plt.show()
```

```python
import numpy as np

# Obtain the number of data points in each cluster
counts = np.bincount(kmeans.labels_)

# Plot the cluster sizes as a histogram
plt.hist(kmeans.labels_)
plt.show()

# Plot the cluster sizes as a bar chart
plt.bar(range(kmeans.n_clusters), counts)
plt.show()
```

```python
# Obtain the cluster assignments of each data point
labels = kmeans.labels_

# Plot the data points with different colors for each cluster
plt.scatter(cluster_data_imputed[:, 0], cluster_data_imputed[:, 1], c=labels)
plt.show()
```

```python
# Calculate the within-cluster sum of squares (WCSS)
wcss = []
for k in range(1, 101):
    kmeans = KMeans(n_clusters=k, n_init=100, random_state=0)
    kmeans.fit(cluster_data_imputed)
    wcss.append(kmeans.inertia_)

# Plot the WCSS
plt.plot(range(1, 101), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

#### Density based clustering

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler, LabelEncoder

# Load the data into a Pandas DataFrame
data = pd.read_csv('./data/athletes.csv')

data = data.dropna()

# Compute a new column with the sum of the weight lifting and timed events
data['total_score'] = data[['fran', 'helen', 'grace', 'filthy50', 'fgonebad', 'run400', 'run5k', 'candj', 'snatch', 'deadlift', 'backsq', 'pullups']].sum(axis=1)

# Convert the gender column to numerical values using label encoding
gender_encoder = LabelEncoder()
data['gender'] = gender_encoder.fit_transform(data['gender'])

# Select the columns to use for clustering
X = data[['gender', 'total_score']]

# Scale the data to have zero mean and unit variance
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform density-based clustering using DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
clusters = dbscan.fit_predict(X_scaled)
```
